```python
from google import genai
from getpass import getpass

api_key = getpass("Enter your Gemini API Key: ")

client = genai.Client(api_key=api_key)
```

```python
print("AI LEAVE LETTER GENERATOR")
print("-" * 40)
```

```python
name = input("Enter student name: ")
student_id = input("Enter student ID: ")
course = input("Enter course/program: ")
email = input("Enter email address: ")
teacher = input("Enter teacher/principal name: ")
start_date = input("Enter leave start date: ")
end_date = input("Enter leave end date: ")
reason = input("Enter reason for leave: ")
```

```python
prompt = f"""
You are an academic communication assistant.

Write a properly formatted formal leave letter for a college student using the information below.

Student Name: {name}
Student ID: {student_id}
Course: {course}
Email: {email}
Teacher Name: {teacher}
Leave Start Date: {start_date}
Leave End Date: {end_date}
Reason: {reason}

Instructions:
- Write the leave letter in a proper formal format.
- Include a subject line.
- Write a detailed body (5–6 sentences).
- Keep each section on separate lines.
- Do NOT include placeholders or square brackets.

Format exactly like this:

Subject: Leave Application

Dear {teacher},

I am writing to request leave from {start_date} to {end_date} due to {reason}. Explain politely why the leave is necessary and mention that the student will catch up on missed classes and assignments.

Thank the teacher for their understanding.

Sincerely,

{name}
Student ID: {student_id}
{course}
Email: {email}
"""
```

```python
from IPython.display import Markdown, display

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print("\nGENERATED LEAVE LETTER")
print("-" * 40)

display(Markdown(response.text))
```
